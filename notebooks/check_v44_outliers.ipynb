```python
for f, al in enumerate([(zout_s, data_s, cat_s, 'goods-s'), (zout_n, data_n, cat_n, 'goods-n')]):
    fig, axes = plt.subplots(1,2, figsize = (15,8))
    zout = al[0]
    data = al[1]
    cat  = al[2]
    field = al[3]



    zs = zout[1].data['z_spec']
    zp = zout[1].data['z_phot']

    gd = where(zs > 0.0)


    axes[0].plot(zs[gd], zp[gd], 'k.')
    axes[0].plot([0,6], [0,6], 'k-')

    axes[0].set_xlim(0,6)
    axes[0].set_ylim(0,6)


    axes[1].plot(zs[gd], zp[gd] - zs[gd], 'k.', markersize = 2)

    x = mad_std(zp[gd] - zs[gd])

    axes[1].axhline(y = 0.0, linestyle = '-')
    axes[1].axhline(y = x, linestyle = '--')
    axes[1].axhline(y = -x, linestyle = '--')

    axes[1].set_ylim(-3*x, 3*x)


    
    sel = where((zs[gd] > 2) & (zs[gd] < 4) & (zp[gd] < 1))[0]
    axes[0].plot(zs[gd][sel], zp[gd][sel], 'rx', markersize = 5)

    
    n_out = len(where(abs(zp[gd] - zs[gd])/(1+zs[gd]) > 0.1)[0])
    n_tot = len(zp[gd])
    axes[0].annotate('f$_{outlier}$ = ' + '%.1f'%(100. * n_out/n_tot)+ r'%' , (0.1, 0.9), xycoords = 'axes fraction', fontsize = 20)
    axes[0].set_title(field, fontsize = 20)
```
