```python
# select galaxies where OII enters G102 (at z=8500./3727 -1 = 1.28) and OIII exits G141 (at z=15000 / 5008 - 1 = 1.99)
def plot_uvj(n, s) : 

    n10z1 = n.df.query('mass >= 1e9 & z_50 >= 1.00 & z_50 <= 2.50 & use_phot == 1')
    s10z1 = s.df.query('mass >= 1e9 & z_50 >= 1.00 & z_50 <= 2.50 & use_phot == 1')

    n10z1q = n10z1.query('UmV > 1.3 & VmJ < 1.5 & UmV >= 0.88*VmJ + 0.59')
    s10z1q = s10z1.query('UmV > 1.3 & VmJ < 1.5 & UmV >= 0.88*VmJ + 0.59')

    # opposite of quiescent selection: 
    n10z1sf = n10z1.query('(UmV < 1.3 | VmJ > 1.5) | UmV <= 0.88*VmJ + 0.59')
    s10z1sf = s10z1.query('(UmV < 1.3 | VmJ > 1.5) | UmV <= 0.88*VmJ + 0.59')

    fig, ax = plt.subplots(figsize=(6,3))
#ax.scatter( -2.5*np.log10(n10z1['restV']/n10z1['restJ']), -2.5*np.log10(n10z1['restU']/n10z1['restV']), s=1)
#ax.scatter( -2.5*np.log10(s10z1['restV']/s10z1['restJ']), -2.5*np.log10(s10z1['restU']/s10z1['restV']), s=1)
    ax.scatter( n10z1q.VmJ, n10z1q.UmV, s=1, c='r', label=None)
    ax.scatter( s10z1q.VmJ, s10z1q.UmV, s=1, c='r', label=None)
    #ax.legend()
    
    ax.scatter( n10z1sf.VmJ, n10z1sf.UmV, s=5, c='b', label='GND 1 < $z_\mathrm{grism}$ < 2.5')
    ax.scatter( s10z1sf.VmJ, s10z1sf.UmV, s=5, c='g', label='GSD 1 < $z_\mathrm{grism}$ < 2.5')


    #colors=100*np.random.rand(len(patches))
    #p.set_array(np.array(colors))
    
    ax.set_xlim(-0.5,2.5)
    ax.set_ylim(-0.5,2.5)
    
    xr = ax.get_xlim()
    yr = ax.get_ylim()
    
    # UmV = 1.3, VmJ = (1.3 - 0.59) / 0.88
    #pg = [[(0.0, 1.3), ((1.3-0.59)/0.88, 1.3), (1.5, 0.88*1.5+0.59), (1.5, 3.0)]]
    pg = [[(xr[0], 1.3), ((1.3-0.59)/0.88, 1.3), (1.5, 0.88*1.5+0.59), (1.5, yr[1])]]
    lc = collections.LineCollection(pg, linewidths=3, color='w')
    ax.add_collection(lc)
    lc = collections.LineCollection(pg, linewidths=1, color='k')
    ax.add_collection(lc)
    

    
    ax.set_xlabel('V - J')
    ax.set_ylabel('U - V')
    ax.legend()
    plt.show()
    plt.savefig('massMetal_uvj.png')
```

```python
# Use Kewley & Dopita 2002 flowchart for R23: 
Z = [0.05, 0.1, 0.2, 0.5, 1.0, 1.5, 2.0, 3.0]
k0 = [-36.9772, -74.2814, -36.7948, -81.1880, -52.6367, -86.8674, -24.4044, 49.4728]
k1 = [10.2838, 24.6206, 10.0581, 27.5082, 16.0880, 28.0455, 2.51913, -27.4711]
k2 = [-0.95742, -2.79194, -0.914212, -3.19126, -1.67443, -3.01747, 0.452486, 4.50304]
k3 = [0.0328614, 0.110773, 0.0300472, 0.128252, 0.0608004, 0.108311, -0.0491711, -0.23228]

qCoeffs = pd.DataFrame( {'Z': Z, 'k0': k0, 'k1': k1, 'k2': k2, 'k3': k3})

def R_ion(x, Z) :
    # x = loq(q)
    
    ZZ = [0.05, 0.1, 0.2, 0.5, 1.0, 1.5, 2.0, 3.0]
    k0 = [-36.9772, -74.2814, -36.7948, -81.1880, -52.6367, -86.8674, -24.4044, 49.4728]
    k1 = [10.2838, 24.6206, 10.0581, 27.5082, 16.0880, 28.0455, 2.51913, -27.4711]
    k2 = [-0.95742, -2.79194, -0.914212, -3.19126, -1.67443, -3.01747, 0.452486, 4.50304]
    k3 = [0.0328614, 0.110773, 0.0300472, 0.128252, 0.0608004, 0.108311, -0.0491711, -0.23228]

    qCoeffs = pd.DataFrame( {'Z': ZZ, 'k0': k0, 'k1': k1, 'k2': k2, 'k3': k3})


    
    df = qCoeffs.iloc[(qCoeffs['Z']-Z).abs().argsort()[:1]]  # see: https://stackoverflow.com/questions/30112202/how-do-i-find-the-closest-values-in-a-pandas-series-to-an-input-number

    #display(df)
    #print(df['k0'].values[0])
    return( df['k0'].values[0] + df['k1'].values[0]*x + df['k2'].values[0]*x**2 + df['k3'].values[0]*x**3)

def q_invert(o2, o3, Z) :
    
    # given the log ratio of OIII/OII:  o3o2 = log10( OIII / OII), return log_q (ionization parameter)
    
    o3o2 = np.log10( o3 / o2)
    
    ZZ = [0.05, 0.1, 0.2, 0.5, 1.0, 1.5, 2.0, 3.0]
    k0 = [-36.9772, -74.2814, -36.7948, -81.1880, -52.6367, -86.8674, -24.4044, 49.4728]
    k1 = [10.2838, 24.6206, 10.0581, 27.5082, 16.0880, 28.0455, 2.51913, -27.4711]
    k2 = [-0.95742, -2.79194, -0.914212, -3.19126, -1.67443, -3.01747, 0.452486, 4.50304]
    k3 = [0.0328614, 0.110773, 0.0300472, 0.128252, 0.0608004, 0.108311, -0.0491711, -0.23228]

    qCoeffs = pd.DataFrame( {'Z': ZZ, 'k0': k0, 'k1': k1, 'k2': k2, 'k3': k3})
    df = qCoeffs.iloc[(qCoeffs['Z']-Z).abs().argsort()[:1]]
    
    myq = np.log10( np.logspace(6.6,8.6,50) ) 
    myO3O2 = [R_ion(x,Z) for x in myq ]
    
    f = interp1d(myO3O2, myq)
    
    #for x,y in zip(myq, myO3O2) : print(x,y)
    print(o3o2)
    return( f(o3o2) )


    
```
