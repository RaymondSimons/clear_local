```python
fig, axes = plt.subplots(1,3, figsize = (25, 10))
xmn = 20
xmx = 60

o2_map = fit_hdu['LINE', 'OII'].data[xmn:xmx, xmn:xmx]
o3_map = fit_hdu['LINE', 'OIII'].data[xmn:xmx, xmn:xmx]
hb_map = fit_hdu['LINE', 'Hb'].data[xmn:xmx, xmn:xmx]





o2_lw = fit_hdu['LINEWHT', 'OII'].data[xmn:xmx, xmn:xmx]
o3_lw = fit_hdu['LINEWHT', 'OIII'].data[xmn:xmx, xmn:xmx]
hb_lw = fit_hdu['LINEWHT', 'Hb'].data[xmn:xmx, xmn:xmx]



vmn = -0.2
vmx = 2.0
axes[0].imshow(o2_map, vmin = vmn, vmax = 1.1)
axes[1].imshow(o3_map, vmin = vmn, vmax = vmx)
axes[2].imshow(hb_map, vmin = vmn, vmax = 0.9)

axes[0].set_title('OII', fontsize = 30)
axes[1].set_title('OIII', fontsize = 30)
axes[2].set_title(r'H$\beta$', fontsize = 30)
```

```python
x = np.arange(xmn - (xmn+xmx)/2, xmx - (xmn+xmx)/2) + 0.5
y = np.arange(xmn - (xmn+xmx)/2, xmx - (xmn+xmx)/2) + 0.5
```

```python
xv, yv = np.meshgrid(x, y)
r = sqrt(xv**2. + yv**2.)
```

```python
fig, axes = plt.subplots(1,3, figsize = (15,4))
axes[0].plot(r.ravel(), o3_map.ravel(),'.', alpha = 0.2, color = 'grey')
axes[1].plot(r.ravel(), o2_map.ravel(),'.', alpha = 0.2, color = 'grey')
axes[2].plot(r.ravel(), hb_map.ravel(),'.', alpha = 0.2, color = 'grey')


bins = linspace(0, r.max(), 20)
bns = mean([bins[0:-1], bins[1:len(bins)]], axis = 0)
fs = zeros((2, 3, len(bins) - 1, 2))

objects = [('GN2', 18197),
           ('GN3', 32719)]





for j in arange(2):
    field = objects[j][0]
    ID = objects[j][1]
    PATH_TO_PREP = glob('/Volumes/gdrive/clear/grizli_extractions/%s/*/Prep'%field)[0]
    fits_file = PATH_TO_PREP + '/%s_%i.full.fits'%(field, ID)
    fit_hdu = fits.open(fits_file)    
    o2_map = fit_hdu['LINE', 'OII'].data[xmn:xmx, xmn:xmx]
    o3_map = fit_hdu['LINE', 'OIII'].data[xmn:xmx, xmn:xmx]
    hb_map = fit_hdu['LINE', 'Hb'].data[xmn:xmx, xmn:xmx]





    o2_lw = fit_hdu['LINEWHT', 'OII'].data[xmn:xmx, xmn:xmx]
    o3_lw = fit_hdu['LINEWHT', 'OIII'].data[xmn:xmx, xmn:xmx]
    hb_lw = fit_hdu['LINEWHT', 'Hb'].data[xmn:xmx, xmn:xmx]

    maps = [(o3_map, o3_lw), (o2_map, o2_lw), (hb_map, hb_lw)]
    for i in arange(3):
        mp, lw = maps[i]
        for b, bn in enumerate(bins[0:-1]):
            good = where((r > bn) & (r < bins[b+1]) & (mp < 10) & (mp > -1))
            fs[j, i, b, 0] = sum(mp[good] * lw[good])/sum(lw[good])
            fs[j, i, b, 1] = sqrt(1./sum(lw[good]))



    #units of line maps are 10^-17 erg/s/cm^2

```

```python
fig, axes = plt.subplots(1,3, figsize = (15,4))
for i in arange(3):
    axes[i].errorbar(bns, fs[0, i,:,0], yerr = fs[0, i,:,1], marker = 'o', linestyle = 'None')

for a, ax in enumerate(axes):
    ax.set_ylim(-0.5, 4)
    ax.axhline(y = 0.0, color = 'grey', linestyle = '--', zorder = 3)
    if a ==0 : ax.set_ylabel(r'line flux (10$^{-17}$ erg s$^{-1}$ cm$^2$)', fontsize = 20)
    ax.set_xlabel(r'pixels', fontsize = 20)
```
