```python
import astropy
from astropy.io import fits
import grizli
from grizli.pipeline import photoz
from grizli import utils
import eazy
import matplotlib.pyplot as plt
import os
from numpy import *
import numpy as np
from astropy.io import ascii
%matplotlib inline


```

```python
#cat_dir = '/Users/rsimons/Desktop/clear/Catalogs'
cat_dir = '/Users/rsimons/Dropbox/rcs_clear/catalogs'
field = 's'

#v41    = ascii.read(cat_dir + '/goods%s_3dhst.v4.1.cats/Catalog/goods%s_3dhst.v4.1.cat'%(field, field))
#v43    = ascii.read(cat_dir + '/goods%s_3dhst.v4.3.cat'%(field))
#v43nzp = ascii.read(cat_dir + '/goods%s_3dhst.v4.3.nzpcat'%(field))
v44 = ascii.read(cat_dir + '/goods%s_3dhst.v4.4.cats/Catalog/goods%s_3dhst.v4.4.cat'%(field))



#translate_file_v41 = '/Users/rsimons/Desktop/clear/Catalogs/goodsn_3dhst.v4.1.translate'
#translate_file_v43 = '/Users/rsimons/Desktop/clear/Catalogs/goodsn_v4.3.translate'
translate_file_v44 = cat_dir + '/goods%s_3dhst.v4.4.cats/Eazy/goodsn_v4.4.translate'

#eazy.symlink_eazy_inputs(path=os.path.dirname(eazy.__file__)+'/data', path_is_env=False)
```

```python
params_v41 = {}
params_v41['CATALOG_FILE'] = cat_dir + '/goods%s_3dhst.v4.1.cats/Catalog/goods%s_3dhst.v4.1.cat'%(field, field)
params_v41['Z_STEP'] = 0.002
params_v41['Z_MAX'] = 4
params_v41['MAIN_OUTPUT_FILE'] = '{0}_3dhst.{1}.eazypy'.format('goodsn', 'v4.1')
params_v41['PRIOR_FILTER'] = 205
params_v41['MW_EBV'] = {'aegis':0.0066, 'cosmos':0.0148, 'goodss':0.0069, 
                    'uds':0.0195, 'goodsn':0.0103}['goodsn']
params_v41['TEMPLATES_FILE'] = 'templates/fsps_full/tweak_fsps_QSF_12_v3.param'




params_v43 = {}
params_v43['CATALOG_FILE'] = cat_dir + '/goods%s_3dhst.v4.3.cat'%(field)
params_v43['Z_STEP'] = 0.002
params_v43['Z_MAX'] = 4
params_v43['MAIN_OUTPUT_FILE'] = '{0}_3dhst.{1}.eazypy'.format('goodsn', 'v4.3')
params_v43['PRIOR_FILTER'] = 205
params_v43['MW_EBV'] = {'aegis':0.0066, 'cosmos':0.0148, 'goodss':0.0069, 
                    'uds':0.0195, 'goodsn':0.0103}['goodsn']
params_v43['TEMPLATES_FILE'] = 'templates/fsps_full/tweak_fsps_QSF_12_v3.param'






params_v44 = {}
params_v44['CATALOG_FILE'] = cat_dir + '/goods%s_3dhst.v4.4.cat'%(field)
params_v44['Z_STEP'] = 0.002
params_v44['Z_MAX'] = 4
params_v44['MAIN_OUTPUT_FILE'] = '{0}_3dhst.{1}.eazypy'.format('goodsn', 'v4.3')
params_v44['PRIOR_FILTER'] = 205
params_v44['MW_EBV'] = {'aegis':0.0066, 'cosmos':0.0148, 'goodss':0.0069, 
                   'uds':0.0195, 'goodsn':0.0103}['goodsn']
params_v44['TEMPLATES_FILE'] = 'templates/fsps_full/tweak_fsps_QSF_12_v3.param'




```

```python
templ0 = grizli.utils.load_templates(fwhm=1200, line_complexes=True, stars=False, 
                                     full_line_list=None,  continuum_list=None, 
                                     fsps_templates=True)

ez_v41 = eazy.photoz.PhotoZ(param_file=None, translate_file=translate_file_v41, 
                        zeropoint_file=None, params=params_v41, 
                        load_prior=True, load_products=False)

ez_v43 = eazy.photoz.PhotoZ(param_file=None, translate_file=translate_file_v41, 
                        zeropoint_file=None, params=params_v43, 
                        load_prior=True, load_products=False)

ez_v44 = eazy.photoz.PhotoZ(param_file=None, translate_file=translate_file_v41, 
                        zeropoint_file=None, params=params_v44, 
                        load_prior=True, load_products=False)

ep_v41 = photoz.EazyPhot(ez_v41, grizli_templates=templ0, zgrid=ez_v41.zgrid)
ep_v43 = photoz.EazyPhot(ez_v43, grizli_templates=templ0, zgrid=ez_v43.zgrid)
ep_v44 = photoz.EazyPhot(ez_v44, grizli_templates=templ0, zgrid=ez_v44.zgrid)
```

```python
di = 16229
ra = v41['ra'][where(v41['id'] == di)[0]]
dec = v41['dec'][where(v41['id'] == di)[0]]

phot_v41, ii, dd = ep_v41.get_phot_dict(float(ra), float(dec))
phot_v43, ii, dd = ep_v43.get_phot_dict(float(ra), float(dec))
phot_v44, ii, dd = ep_v44.get_phot_dict(float(ra), float(dec))

fig, ax = plt.subplots(1,1, figsize = (10,10))
ax.set_xlabel('$\lambda$')
ax.set_ylabel('f$_{\lambda}$')
clrs = ['black', 'red', 'blue']
mrks = ['D', 'x', 'o']
for p, phot in enumerate([phot_v41, phot_v43, phot_v44]):
    for i in arange(len(phot['filters'])):
        wave = phot['filters'][i].pivot() * 1.e-4
        flam = phot['flam'][i] / 1.e-19
        eflam = phot['eflam'][i] / 1.e-19
        if flam > 0:
            ax.errorbar(wave, flam, eflam, fmt = 'o', color = clrs[p], marker = mrks[p])
        print (phot['filters'][i].name)
        print ('%i   %.1f   %.6f'%(i, flam, wave))
ax.set_xscale('log')
ax.set_xlim(0.3, 10)
#ax.set_ylim(0, 130)

fig.savefig('test.png')



```

```python

def magn(f):
    return 25 - 2.5 * log10(f)




fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize = (5, 15))


for (ax, flt1, flt2) in [(ax1, 'f_Z', 'f_F850LP'), (ax2, 'f_R', 'f_F606W'), (ax3, 'f_I','f_F775W' )]:
    try:
        
        mag_z_41 = magn(v41[flt1])
        mag_z_43 = magn(v43[flt1])
        mag_z_44 = magn(v44[flt1])

        mag_8_41 = magn(v41[flt2])
        mag_8_43 = magn(v43[flt2])
        mag_8_44 = magn(v44[flt2])

        gd = where(mag_8_41 < 25)


        pl = ax.hist(mag_z_41[gd] - mag_8_41[gd], bins = linspace(-0.75, 0.75, 100), histtype = 'step', color = 'black', label = 'v41')
        pl = ax.hist(mag_z_43[gd] - mag_8_43[gd], bins = linspace(-0.75, 0.75, 100), histtype = 'step', color = 'red', label = 'v43')
        pl = ax.hist(mag_z_44[gd] - mag_8_44[gd], bins = linspace(-0.75, 0.75, 100), histtype = 'step', color = 'blue', label = 'v44')


        ax.set_xlabel('%s - %s'%(flt1.replace('f_', 'm_subaru,'), flt2.replace('f_', 'm_HST,')))



        ax.axvline(x = 0, color = 'grey')

        ax.legend()
    except:
        pass

fig.savefig('/Users/rsimons/Desktop/clear/figures/comparison.png', dpi = 300)
```
