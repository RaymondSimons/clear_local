```python
%matplotlib inline
import astropy
from astropy.io import fits
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from  matplotlib.pyplot import *

```

```python
def write_v44_catalog(field, cat_fname, header_str, zp_filters, translate, zp_corrections, catalog_orig, cat_plus):
    cat_new = catalog_orig.copy()
    print (shape(cat_new))
    f = open(cat_fname, 'w+')
    
    colnames = array(header_str.lstrip('# ').split(' '))
    
    for ff, filt in enumerate(zp_filters):
        t_gd = where(filt == translate[:,1])[0][0]
        translate_filt = translate[t_gd,0]        
        c_gd = where(colnames == translate_filt)[0][0]
        zp = zp_corrections[ff]
        gd = where(catalog_orig[:,c_gd].astype('float') !=-99)[0]
        cat_new[gd,c_gd] = array(['%.8f'%(float(m) * zp) for m in catalog_orig[gd,c_gd]])
        cat_new[gd,c_gd+1] = array(['%.8f'%(float(m) * zp) for m in catalog_orig[gd,c_gd+1]])

    for i in arange(len(cat_new), len(cat_plus)):
        if field == 's': temp_arr = cat_plus[i, 0:shape(cat_new)[1]]
        elif field == 'n': temp_arr = concatenate([cat_plus[i, 0:shape(cat_new)[1]], array(['-99', '-99', '-99', '-99'])])
        cat_new = concatenate([cat_new, array([temp_arr])], axis = 0)
        
    f.write(header_str + '\n# goods%s_3dhst.v4.4.cat\n'%field)

    
    for i in arange(cat_new.shape[0]):
        for j in arange(cat_new.shape[1]):
            f.write(cat_new[i,j] + ' ')
        f.write('\n')


        
    f.close()
    
    return
```

```python
cat_dir = '/Users/rsimons/Desktop/clear/Catalogs'
for field in ['n', 's']:

    zp_corrections = np.loadtxt(cat_dir + '/goods%s_t_3dhst.v4.4.zphot.zeropoint'%field, usecols = (1))
    zp_filters = np.loadtxt(cat_dir + '/goods%s_t_3dhst.v4.4.zphot.zeropoint'%field, dtype = 'str', usecols = (0))
    
    translate      = np.loadtxt(cat_dir + '/goods%s_t_3dhst.v4.4.zphot.translate'%field, dtype = 'str', usecols = (0, 1))
    catalog_orig   = np.loadtxt(cat_dir + '/goods%s_3dhst_temp.v4.3.cat'%field, dtype = 'str')  
    cat_plus       = np.loadtxt(cat_dir + '/Goods%s_plus.cat'%field.upper(), dtype = 'str') 
    header_str     = np.loadtxt(cat_dir + '/goods%s_3dhst_temp.v4.3.cat'%field, comments = '%', dtype = 'str', delimiter = 'asdalksd')[0]
    cat_fname      = '/Users/rsimons/Desktop/clear/Catalogs/goods%s_3dhst.v4.4.cat'%field



    write_v44_catalog(field, cat_fname, header_str, zp_filters, translate, zp_corrections, catalog_orig, cat_plus)

```

```python
for f, field in enumerate(array(['n', 's'])):
    cat_old = np.loadtxt('/Users/rsimons/Desktop/clear/Catalogs/goods%s_3dhst.v4.1.cats/Catalog/goods%s_3dhst.v4.1.cat'%(field, field))
    cat_new = np.loadtxt('/Users/rsimons/Desktop/clear/Catalogs/goods%s_3dhst.v4.4.cat'%field)    
    header_str = np.loadtxt('/Users/rsimons/Desktop/clear/Catalogs/goods%s_3dhst_temp.v4.3.cat'%field, comments = '%', dtype = 'str', delimiter = 'asdalksd')[0]
    colnames = array(header_str.lstrip('# ').split(' '))
    zp_corrections = np.loadtxt(cat_dir + '/goods%s_3dhst.v4.4.zphot.zeropoint'%field, usecols = (1))
    zp_filters = np.loadtxt(cat_dir + '/goods%s_3dhst.v4.4.zphot.zeropoint'%field, dtype = 'str', usecols = (0))
    translate      = np.loadtxt(cat_dir + '/goods%s_3dhst.v4.4.zphot.translate'%field, dtype = 'str', usecols = (0, 1))

    
    fig, axes = plt.subplots(len(zp_filters), 2, figsize = (8, 4*len(zp_filters)))
    for ff, filt in enumerate(zp_filters):
        zp = zp_corrections[ff]
        if filt != 'F202':
            t_gd = where(filt == translate[:,1])[0][0]
            translate_filt = translate[t_gd,0]        
            c_gd = where(colnames == translate_filt)[0][0]            
            fold = cat_old[:,c_gd]
            fnew = cat_new[0:len(cat_old),c_gd]
            eold = cat_old[:,c_gd+1]
            enew = cat_new[0:len(cat_old),c_gd+1]
            axes[ff,0].plot(fold, fnew/fold,'k.')
            axes[ff,1].plot(fold, enew/eold,'k.')
            axes[ff,0].annotate(filt, (0.1, 0.9), xycoords = 'axes fraction')            
            axes[ff,0].set_ylabel('f_v44/f_v41')
            axes[ff,0].set_xlabel('f_v41')
            axes[ff,1].set_ylabel('e_v44/e_v41')
            axes[ff,1].set_xlabel('f_v41')
            
            for ax in axes[ff]:
                ax.axhline(y = zp, color = 'red', linestyle = '-')
                ax.set_ylim(0.90, 1.10)
                ax.set_xscale('log')
                ax.axhline(y = 1, color = 'grey', linestyle = '--')
    fig.subplots_adjust(wspace = 0.2)
    fig.savefig(cat_dir + '/v44_v41_GOODS-%s.png'%field.upper(), dpi = 300)
    
    
    
    
    
    
```
